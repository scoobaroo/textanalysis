```python
import pandas
import numpy
import scipy
import string
from string import digits
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

def splitBookIntoParagraphs(filename):
    file = open(filename, "r")
    lines = file.read()
    array = lines.split('\n\n')
    del array[0]
    del array[0]
    newArray = []
    for string in array:
        endIndex = len(string)
        newstring = string.replace('\n', ' ').lower()
        newArray.append(newstring)
    return newArray

conneticutyankee = splitBookIntoParagraphs("./marktwain/marktwain_aconneticutyankee.txt")
print(conneticutyankee)

```

```python
import nltk
from collections import Counter
#pass in a tags object
def getcounts(taggedtext):
    counts = Counter(tag for word,tag in taggedtext)
    total = sum(counts.values())
    return dict((word, float(count)/total) for word,count in counts.items())
CYcountsArray = []
for paragraph in conneticutyankee:
    tokens = nltk.word_tokenize(paragraph)
    text = nltk.Text(tokens)
    tags = nltk.pos_tag(text,tagset="universal")
    counts = getcounts(tags)
    global countsArray
    CYcountsArray.append(counts)
    
print(CYcountsArray)

count = 0
def findtags(tag_prefix, tagged_text):
    cfd = nltk.ConditionalFreqDist((tag, word) for (word, tag) in tagged_text if tag.startswith(tag_prefix))
    for(word,tag) in tagged_text:
        if tag.startswith(tag_prefix):
            global count
            count = count + 1
    return dict((tag, cfd[tag].most_common(10)) for tag in cfd.conditions())


```

```python
#here we will import the other 4 books from each author for training; we'll do mark twain first
def GetPOSCounts(array):
    countsArray = []
    for paragraph in array:
        tokens = nltk.word_tokenize(paragraph)
        text = nltk.Text(tokens)
        tags = nltk.pos_tag(text,tagset="universal")
        counts = getcounts(tags)
        countsArray.append(counts)
    return countsArray
 
huckleberryfinn = splitBookIntoParagraphs('./marktwain/marktwain_adventuresofhuckleberryfinn.txt')
huckleberryfinn
```

```python
import unicodedata
newArr = []
for string in huckleberryfinn:
    global newArr
    newstring = string.replace(u'\xa0' , u'')
    newArr.append(newstring)
    
huckleberryfinn = newArr
```

```python
hfArray = GetPOSCounts(huckleberryfinn)
hfTrainX = pd.DataFrame(hfArray)
hfTrainX = hfTrainX.fillna(0)
def makeData(array):
    array = pd.DataFrame(array)
    del array['X']
    array = array.fillna(0)
    return array
```
